```python
%sql
USE CATALOG etl_project;
USE SCHEMA bigquery_db_cohort_db;

CREATE OR REPLACE TABLE gold_repeat_purchases AS
SELECT
    cohort_month,
    ROUND(
        1.0 * COUNT_IF(total_orders >= 2) / COUNT(*),
        4
    ) AS repeat_rate_2_orders,
    ROUND(
        1.0 * COUNT_IF(total_orders >= 3) / COUNT(*),
        4
    ) AS repeat_rate_3_orders,
    ROUND(
        1.0 * COUNT_IF(total_orders >= 4) / COUNT(*),
        4
    ) AS repeat_rate_4_orders
FROM silver_customer_facts
GROUP BY cohort_month
ORDER BY cohort_month;
```
