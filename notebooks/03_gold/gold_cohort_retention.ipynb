```python
%sql
USE CATALOG etl_project;
USE SCHEMA bigquery_db_cohort_db;

CREATE OR REPLACE TABLE gold_cohort_retention AS
WITH base AS (
    SELECT
        cohort_month,
        months_between
    FROM silver_customer_facts
)
SELECT
    cohort_month,
    ROUND(
        1.0 * COUNT_IF(months_between = 0) / COUNT(*),
        4
    ) AS retention_1m,
    ROUND(
        1.0 * COUNT_IF(months_between IN (0, 1)) / COUNT(*),
        4
    ) AS retention_2m,
    ROUND(
        1.0 * COUNT_IF(months_between IN (0, 1, 2)) / COUNT(*),
        4
    ) AS retention_3m
FROM base
GROUP BY cohort_month
ORDER BY cohort_month;
```
